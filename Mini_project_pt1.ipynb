## Explore and clean data

```python
df.info() #show the data types and number of null values 
```

```python
#rename columns
df = df.rename(columns={"Crm Cd Desc": "Crime", "DR_NO": "ID"}) 
df.head()
```

```python
df.isnull().sum() #show the number of rows containing null value
```

```python
df = df.dropna() #drop null values
df.isnull().sum()
```

```python
df.describe() #statistical summary of the data
```

```python
df = df[(df['Crm Cd']>=0) & (df['Vict Age']>=0) & (df['AREA']>=0)] #filter data
df.describe()
```

```python
df['Crm Cd'].value_counts() #show the count of unique values 
```

```python
df = df[['Vict Age','Crm Cd','Crm Cd 1','Crm Cd 2','Crm Cd 3','Crm Cd 4','AREA','TIME OCC']] # keep specific columns 
df.head()
```

```python
df.set_index('Vict Age',inplace=True) #define the index
df.head()
```

```python
df.to_csv('clean_crime_data.csv') #store the cleaned data to local computer
```

## Visualize data

### Distribution of the Data

```python
df.hist(layout=(3,3), figsize=(20, 10), sharey=False, sharex=False, bins=50) #use multiple charts to show the distribution of all columns
```

### The Sum of Crimes per Age

```python
sum_crimes_by_age = df.groupby('Vict Age')['Crm Cd'].sum() #aggregate the total case number per day
sum_crimes_by_age.head()
```

```python
sum_crimes_by_age.plot() #use a line chart to show the result
```

### The Top Areas with the Highest Deaths

```python
sum_crimes_by_AREA = df.groupby('AREA')['Crm Cd'].sum() # calculate the total deaths per day
sum_crimes_by_AREA.nlargest(10).plot.bar() # use a bar chart to show the results of the top 10 countries 
```

### Victem Age Crime  Visualization

```python
AREA_19_data = df[df['AREA']=='19'] #select the USA data
```

```python
AREA_19_data.reset_index().plot.scatter(x='Vict Age',y='TIME OCC',c='Crm Cd') #use scatter plot to show the result
```
